```python
import pandas
```

# 0. user-defined variables

```python
identifiers_file = '/home/adrian/projects/ATG7/data/samples_primary_tumors.tsv'
storing_file = '/home/adrian/projects/ATG7/results/primary_tumors_expression.tsv'

storing_file_1000samples_1000genes = '/home/adrian/projects/ATG7/results/primary_tumors_expression.1e3s.1e3g.tsv'
storing_file_allsamples_1000genes = '/home/adrian/projects/ATG7/results/primary_tumors_expression.alls.1e3g.tsv'
storing_file_1000samples_allgenes = '/home/adrian/projects/ATG7/results/primary_tumors_expression.1e3s.allg.tsv'

expression_file = '/home/adrian/projects/ATG7/data/TcgaTargetGtex_rsem_gene_tpm.txt'
```

```python
# 1. read data
```

```python
organ_IDs = pandas.read_csv(identifiers_file, sep='\t')
organ_IDs.head()
```

```python
organ_labels = organ_IDs['sample'].to_list()
print(len(organ_labels))
print(organ_labels[:10])
```

```python
%%time
expression = pandas.read_csv(expression_file, sep='\t', index_col='sample')
print(expression.shape)
expression.head()
```

# 2. retrieve expression of sample identifiers

```python
subexpression = expression[organ_labels]
```

```python
print(subexpression.shape)
```

# 3. store data

```python
%%time
subexpression.to_csv(storing_file, sep='\t')
```
